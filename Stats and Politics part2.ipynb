> ### __3.3__ Look back at the non-aggregated county-level data. Conduct one or more statistical tests to see whether the mean Republican party vote share across all of the counties are different for the 3 elections (Which statistical test is appropriate here? Assume that the people who voted in each county are a representative sample of its population.) Describe why you’ve chosen your tests and interpret your results.

>> We are going to use two sample t-test to find out whether the means of Republican party vote share over all counties are different for three elections. We choose this test because is used to test the difference between two population means and a common application is to determine whether the means are equal.

>> From previous we have a dataframe contains counties and Republican party vote share called *county_df*.  


>> We define a *calculateTtest* function to have an easier way to calculate p-value and t-test:

```python
def calculateTtest(a, b):
    t_test, p_val = stats.ttest_ind(a, b)
    
    if p_val < 0.05:
        print("We can reject null hypothesis")
    else:
        print("We accept null hypothesis")
        
    return t_test, p_val
```

>> Comparison between Republican vote share in 2008 and 2012:


```python
ttest, pval = calculateTtest(county_df['Republicans 12 pct'], county_df['Republicans 08 pct'])
print('t-test:', ttest)
print('p-value: ', pval)
```

>> Comparison between Republican vote share in 2008 and 2016:


```python
ttest, pval = calculateTtest(county_df['Republicans 16 pct'], county_df['Republicans 08 pct'])
print('t-test:', ttest)
print('p-value: ', pval)
```

>> Comparison between Republican vote share in 2012 and 2016:


```python
ttest, pval = calculateTtest(county_df['Republicans 16 pct'], county_df['Republicans 12 pct'])
print('t-test:', ttest)
print('p-value: ', pval)
```
